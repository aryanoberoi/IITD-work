```python
import pandas as pd
import torch
from transformers import BertTokenizer, BertModel

json_file_path = 'Books.json'

df = pd.read_json(json_file_path, lines=True, nrows=10000)

df.dropna(subset=['reviewText'], inplace=True)


print(df)
```

```python
import pandas as pd
import torch
from transformers import BertTokenizer, BertModel

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')


def get_bert_embedding(text):
    
    tokens = tokenizer.encode(text, add_special_tokens=True, max_length=512, truncation=True)

    # Convert tokens to tensor
    input_ids = torch.tensor(tokens).unsqueeze(0)

    # getting hidden states
    with torch.no_grad():
        outputs = model(input_ids)

    # Embedding of CLS token
    cls_embedding = outputs.last_hidden_state[:, 0, :]

    return cls_embedding.numpy()


df['BERT_Embedding'] = df['reviewText'].apply(get_bert_embedding)

print(df)
```
